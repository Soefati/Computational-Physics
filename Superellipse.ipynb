```python
# Libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sp

# Initialize and Declaration
c = 1
n = 20
```

```python
# Visualization
x = np.linspace(-c, c, 1000)
y = np.linspace(-c, c, 1000)
X, Y = np.meshgrid(x, y)
for i in range(2, n + 1, 2):
    F = X ** i + Y ** i - c # X^n + Y^n = c
    plt.contour(X, Y, F, [0], colors = 'orange')
plt.axis('equal')
plt.xlim(-2*c, 2*c)
plt.ylim(-2*c, 2*c)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Superellipse Visualization")
plt.show()
```

```python
for i in range(2, n + 1, 2):
    A = sp.gamma(1 + 1/i) ** 2 / sp.gamma(1 + 2/i)
    plt.scatter(i, A, color = 'blue')
plt.xlim(0, n)
plt.xlabel("n")
plt.ylabel("A")
plt.title("Superellipse Area Visualization")
plt.show()
```
