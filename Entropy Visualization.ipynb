```python
# Libraries
import matplotlib.pyplot as plt
import numpy as np

# Initialize and Declaration
exp, pi, cos, sin, log = np.exp, np.pi, np.cos, np.sin, np.log
ratioTemp, ratioPress, ratioVolum = 10, 10, 10
Cv, Cp, R = 1, 1, 8.3145
T, P, V = np.linspace(1, ratioTemp, 1000), np.linspace(1, ratioPress, 1000), np.linspace(1, ratioVolum, 1000)
```

```python
# Calculation
def dS(x, y):
    return Cv * log(x) + R * log(y) # dS(T, V)
X, Y = np.meshgrid(T, V)
Z = dS(X, Y)
```

```python
# Visualization 3D
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.plot_surface(X, Y, Z, cmap = "hot")
ax.grid(False)
ax.set_xlabel("T")
ax.set_ylabel("V")
ax.set_zlabel("dS")
plt.show()
```

```python
# Visualization 2D
Z = Z[:-1, :-1]
fig, ax = plt.subplots()
plt.pcolormesh(X, Y, Z, cmap = "hot")
plt.xlabel("T")
plt.ylabel("V")
plt.colorbar()
plt.show()
```
