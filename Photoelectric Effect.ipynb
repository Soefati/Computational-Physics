```python
# Libraries
import matplotlib.pyplot as plt
import numpy as np

# Initialize and Declaration
h = 6.626e-34 # Plank's Constant
f = np.linspace(0, 1e15, 1000) # Frequency
w = 3e-19 # Work Function
```

```python
# Calculation
E = lambda f: h * f - w
```

```python
# Visualization
plt.title("Visualization of Kinetic Energy")
plt.plot(f, E(f), color = 'blue')
plt.xlabel("Frequency Photon [Hz]")
plt.ylabel("Kinetics Energy [J]")
plt.ylim(0, np.amax(E(f)))
plt.show()
```
