```python
# Libraries
import matplotlib.pyplot as plt
import numpy as np

# Initialize and Declaration (Data)
pi = np.pi
angle = [ 0, pi * 1 / 3, pi * 2 / 3, pi, pi * 4 / 3, pi * 5 / 3, pi * 2 ] # Horizontal Axis
torque = [ 0, 1066, -323, 0, 323, -355, 0 ] # Vertical Axis
n = len(angle)
matrix = np.array([[1, 0, 0, 0, 0, 0, 0],
                    [1, angle[1], angle[1] ** 2, angle[1] ** 3, angle[1] ** 4, angle[1] ** 5, angle[1] ** 6],
                    [1, angle[2], angle[2] ** 2, angle[2] ** 3, angle[2] ** 4, angle[2] ** 5, angle[2] ** 6],
                    [1, angle[3], angle[3] ** 2, angle[3] ** 3, angle[3] ** 4, angle[3] ** 5, angle[3] ** 6],
                    [1, angle[4], angle[4] ** 2, angle[4] ** 3, angle[4] ** 4, angle[4] ** 5, angle[4] ** 6],
                    [1, angle[5], angle[5] ** 2, angle[5] ** 3, angle[5] ** 4, angle[5] ** 5, angle[5] ** 6],
                    [1, angle[6], angle[6] ** 2, angle[6] ** 3, angle[6] ** 4, angle[6] ** 5, angle[6] ** 6]])
constant = np.array([[0], [1066], [-323], [0], [323], [-355], [0]])
det_matrix = np.linalg.det(matrix)
ordo = len(matrix)
matrix_var = np.zeros((ordo, 1))
```

```python
det_matrix
```

```python
# Calculation (Cramer's Rule)
for i in range(0, ordo):
    for j in range(0, ordo):
        matrix[j, i] = constant[j][0]
    det_matrix_i = np.linalg.det(matrix)
    matrix_var[i][0] = det_matrix_i / det_matrix
    matrix = np.array([[1, 0, 0, 0, 0, 0, 0],
                    [1, angle[1], angle[1] ** 2, angle[1] ** 3, angle[1] ** 4, angle[1] ** 5, angle[1] ** 6],
                    [1, angle[2], angle[2] ** 2, angle[2] ** 3, angle[2] ** 4, angle[2] ** 5, angle[2] ** 6],
                    [1, angle[3], angle[3] ** 2, angle[3] ** 3, angle[3] ** 4, angle[3] ** 5, angle[3] ** 6],
                    [1, angle[4], angle[4] ** 2, angle[4] ** 3, angle[4] ** 4, angle[4] ** 5, angle[4] ** 6],
                    [1, angle[5], angle[5] ** 2, angle[5] ** 3, angle[5] ** 4, angle[5] ** 5, angle[5] ** 6],
                    [1, angle[6], angle[6] ** 2, angle[6] ** 3, angle[6] ** 4, angle[6] ** 5, angle[6] ** 6]])
```

```python
m = matrix_var
a = round(m[6][0], 4)
b = round(m[5][0], 4)
c = round(m[4][0], 4)
d = round(m[3][0], 4)
e = round(m[2][0], 4)
f = round(m[1][0], 4)
g = round(m[0][0], 4)
print("\nSolution:")
print(m)
print("or")
print(f'fx = {a}x^6 + {b}x^5 + {c}x^4 + {d}x^3 + {e}x^2 + {f}x + {g}')
```

```python
x = np.linspace(min(angle), max(angle), 1000)
y = np.zeros(len(x))
fx = (a * x ** 6) + (b * x ** 5) + (c * x ** 4) + (d * x ** 3) + (e * x ** 2) + (f * x) + g
plt.title("Visualization of Polynomial Functions")
plt.scatter(angle, torque, color = "gray", label = "Data")
plt.plot(x, fx, color = "black", label = "Polynomial Functions")
plt.xlabel("Angle (Rad)")
plt.ylabel("Torque (Nm)")
plt.legend(loc = "best")
plt.show()
```
