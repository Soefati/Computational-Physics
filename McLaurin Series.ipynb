MacLaurin Series

```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialize and Declaration
orde = 7
```

For Sine(x)

```python
# Sine(x) Visualization
x = np.linspace(-np.pi, np.pi, 250)
y = np.zeros(len(x))
plt.figure()
plt.plot(x, np.sin(x), "k", label = "Analytic")
for n in range(orde + 1):
    y += (((-1) ** n) * ((x) ** (2 * n + 1))) / (np.math.factorial(2 * n + 1))
    plt.plot(x, y, label = n)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("MacLaurin Series Visualization")
plt.legend(loc = "best")
plt.show()
```

For Cosine(x)

```python
# Cosine(x) Visualization
x = np.linspace(-np.pi, np.pi, 250)
y = np.zeros(len(x))
plt.figure()
plt.plot(x, np.cos(x), "k", label = "Analytic")
for n in range(orde + 1):
    y += (((-1) ** n) * ((x) ** (2 * n))) / (np.math.factorial(2 * n))
    plt.plot(x, y, label = n)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("MacLaurin Series Visualization")
plt.legend(loc = "best")
plt.show()
```

Exp(x)

```python
# Cosine(x) Visualization
x = np.linspace(-np.pi, np.pi, 250)
y = np.zeros(len(x))
plt.figure()
plt.plot(x, np.exp(x), "k", label = "Analytic")
for n in range(orde + 1):
    y += x ** n / (np.math.factorial(n))
    plt.plot(x, y, label = n)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("MacLaurin Series Visualization")
plt.legend(loc = "best")
plt.show()
```
