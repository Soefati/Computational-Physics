```python
# Libraries
from mpmath import*
import numpy as np
import matplotlib.pyplot as plt

# Initialize and Declaration
pi, exp = np.pi, np.exp
c = 299792458
h = 6.626e-34
kb = 1.381e-23
```

```python
# Calculation
def RayleighJeans(f, T): # Rayleigh-Jeans Formula
    return 8 * pi * f ** 2 * kb * T / c ** 3
def Planck(f, T): # Planck Radiation Formula
    return 8 * pi * h * f ** 3 / (c ** 3 * exp(h * f / (kb * T)) - 1)
```

```python
# Visualization
x = np.linspace(0, 2 * 10 ** 15, 1000)
for i in range (3000, 8000, 1000):
    y = lambda f: Planck(f, i)
    plt.plot(x, y(x), label = f'Planck(T) = {i}K')
for i in range (3000, 8000, 1000):
    z = lambda f: RayleighJeans(f, i)
    plt.plot(x, z(x), label = f'Rayleigh-Jeans(T) = {i}K')
plt.plot
plt.xlabel("f")
plt.ylabel("E(f)")
plt.xlim(0, 2 * 10 ** 15)
plt.ylim(0, 3 * 10 ** (-15))
plt.title("Visualization of Radiation Energy Distribution")
plt.legend(loc = "best")
plt.show()
```
