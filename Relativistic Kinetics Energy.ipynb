```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialize and Declaration
m = 1
c = 299792458
v = np.linspace(0, c * 0.99, 1000)
```

```python
# Calculation
KENon = lambda v: 0.5 * m * v ** 2
KE = lambda v: m * c ** 2 *(1 / (1 - v ** 2 / c ** 2) ** 0.5 - 1)
```

```python
# Visualization
plt.plot(v, KENon(v), 'gray', label = 'Classical Kinetics Energy')
plt.plot(v, KE(v), 'blue', label = 'Relativistic Kinetics Energy')
plt.legend(loc = 'best')
plt.xlabel('Velocity')
plt.ylabel('Kinetics Energy')
plt.show()
```
