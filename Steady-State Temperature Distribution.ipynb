Steady-state temperature distribution for the semi-infinite plate problem if the temperature of the bottom edge is T=f(x)=x◦ (in degrees; that is, the temperature at x cm is x degrees), the temperature of the other sides is 0◦, and the width of the plate is 10 cm.

```python
# Libraries
from scipy import integrate
import matplotlib.pyplot as plt
import numpy as np

# Initialize and Declaration
exp, pi, cos, sin = np.exp, np.pi, np.cos, np.sin
length, orde = 10, 200
x = y = xn = np.arange(0, length, .01)
```

```python
# Calculation
bn = lambda n: 2 * (integrate.simpson(xn * sin(n*pi*xn/length), xn)) / length # Fourier Coefficient
def temperature(x, y):
    return sum([bn(j) * exp(-j*pi*y/length) * sin(j*pi*x/length)  for j in range(1, orde + 1)])
X, Y = np.meshgrid(x, y)
Z = temperature(X, Y)
```

```python
# Visualization 3D
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.plot_surface(X, Y, Z, cmap = "hot")
ax.grid(False)
ax.set_title("3D Visualization of Temperature Distribution")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Temperature")
plt.show()
```

```python
# Visualization 2D
Z = Z[:-1, :-1]
fig, ax = plt.subplots()
ax.set_title("2D Visualization of Temperature Distribution")
plt.pcolormesh(X, Y, Z, cmap = "hot")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar()
plt.show()
```
