Legendre Series

```python
# Libraries
from scipy import integrate
import matplotlib.pyplot as plt
import numpy as np

# Initialize and Declaration
orde = 20
```

```python
# Calculation
def P(l, x): # Legendre Polynomial Recursive Functions
    if (l == 0):
        return 1
    elif (l == 1):
        return x
    else:
        return ((2 * l - 1) * x * P(l - 1, x) - (l - 1) * P(l - 2, x)) / float(l)
def c(n): # Coefficient
    y1 = lambda x: (.5 * x ** 2 - 2) * P(n, x)
    y2 = lambda x: 0 * P(n, x)
    integration = integrate.quad(y1, -1, 0)[0] + integrate.quad(y2, 0, 1)[0]
    return .5 * (2 * n + 1) * integration
for i in range(0, orde + 1):
    print(f'c{i} = {round(c(i), 5)}')
```

```python
# Visualization
x = np.arange(-1, 1, .01)
y = lambda x: np.array([.5 * i ** 2 - 2 if -1<i<0 else 0 for i in x])
plt.plot(x, y(x), color = "blue", label = "Analytic")
for i in range (1, orde + 1, 4):
    series = sum([c(j) * P(j, x) for j in range(0, i + 1)])
    plt.plot(x, series, label = "Legendre Series: n = " + str(i))
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Legendre Series Visualization")
plt.legend(loc = "best")
plt.show()
```
